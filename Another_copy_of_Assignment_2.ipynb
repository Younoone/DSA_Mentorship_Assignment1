````python
FF = pd.read_csv("fastfood.csv")
````

````python
FF.head()
````

````python
FF.isnull().sum()
````

````python
FF['fiber'] = FF['fiber'].fillna(FF['fiber'].mean)
````

````python
FF['vit_a'] = FF['vit_a'].fillna(FF['vit_a'].mean)
````

````python
FF['vit_c'] = FF['vit_c'].fillna(FF['vit_c'].mean)
````

````python
FF['calcium'] = FF['calcium'].fillna(FF['calcium'].mean)
````

````python
FF['protein'] = FF['protein'].fillna(FF['protein'].mean)
````

````python
FF.isnull().sum()
````

````python
FF.head()
````

````python
FF['Healthy'].unique()
````

````python
(FF['Healthy']==80).sum()
````

````python
#FF.loc[FF['Healthy']=="80","Healthy"]="1"
#FF['Healthy'].mask(FF['Healthy'] == '80', 1, inplace=True)
````

````python
FF['Healthy'] = FF['Healthy'].replace(80, 1)
````

````python
FF['Healthy'].unique()
````

````python
FF.corr()
````

**Split Dataset into Dependent and Independent set**

````python
X = FF[['calories','cal_fat','total_fat','sat_fat','trans_fat','cholesterol','sodium','total_carb','sugar']]
y = FF['Healthy']
````

````python
X.head()
````

````python
y.head()
````

````python
from sklearn.model_selection import train_test_split
````

````python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
````

````python
X_train.shape
````

````python
X_test.shape
````

````python
y_train.shape
````

````python
y_test.shape
````

````python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
````

````python
from sklearn.model_selection import GridSearchCV
parameter = {'penalty':['l1','l2','elasticnet'],'C':[1,2,3,4,5,6,10,20,30,40,50,60],'max_iter':[100,200]}
````

````python
classifier_regressor=GridSearchCV(classifier,param_grid=parameter,scoring='accuracy',cv=5)
````

````python
X_train = X_train.to_numpy()
y_train = y_train.to_numpy()
````

````python
classifier_regressor.fit(X_train, y_train)
````

````python
print(classifier_regressor.best_params_)
````

````python
print(classifier_regressor.best_score_)
````

````python
##Predict
y_pred=classifier_regressor.predict(X_test)
````

````python
##Accuracy
from sklearn.metrics import accuracy_score,classification_report
````

````python
score=accuracy_score(y_pred,y_test)
print(score)
````

````python
print(classification_report(y_pred,y_test))
````
